## **Operadores Lógicos**

# **Negación**

```python
# Definicion de la proposición original
calor = True

# Negación de la proposición
negacion = not calor

# Imprimir los resultados

print(f"Proposición original. Hace calor: {calor}")
print(f"Negación de la proposición original. Hace calor: {negacion}")
```

# **Conjunción (∧) - "Y"**

Ambas deben ser verdaderas para que toda la oración sea verdadera.

```python
# Proposición 1
hoy_llueve = True

# Proposición 2
hace_frio = True

frase = hoy_llueve and hace_frio

print(frase)
```

# **Disyunción (∨) - "O"**

Basta que una sea verdadera para que todo sea verdadero.

```python
# Definición de las proposiciones

voy_cine = True
no_voy_cine = False

# Disyunción de las proposiciones

frase = voy_cine or no_voy_cine

print(frase)
```

```python
menor_12 = True
estudiante = False

frase = menor_12 or estudiante

print(frase)
```

**El Acceso a la Montaña Rusa**

Para subir, hay reglas estrictas:
Seguridad: Medir más de 140 cm Y no tener problemas cardíacos.Vía Rápida: Si tienes el "Pase VIP", entras sin importar lo anterior (pero siempre que midas más de 120 cm por seguridad básica).

```python
estatura = int(input("¿Cuál es tu estatura? (en cm): "))
salud = str(input("¿Tienes problemas cardíacos? (si/no): ")).lower().strip()
pase_vip = str(input("¿Tienes el pase VIP? (si/no): ")).lower().strip()

if estatura >=140 and salud == "no":
  print("Puedes subir a la montaña rusa")

elif estatura >=120 and pase_vip == "si":
  print("Puedes subir a la montaña rusa")

elif estatura <120 or salud == "no":
  print("No puedes subir a la montaña rusa")

else:
  print("No puedes subir a la montaña rusa")

```
